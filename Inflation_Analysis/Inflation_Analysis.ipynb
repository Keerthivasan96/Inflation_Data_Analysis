 Global Inflation Trend Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example countries for trend plot
countries = ["India", "United States", "China", "Brazil", "Germany"]
df_countries = df_merge[df_merge["Country Name"].isin(countries)]

plt.figure(figsize=(14, 7))
for country in countries:
    country_data = df_countries[df_countries["Country Name"] == country]
    plt.plot([int(col) for col in year_columns], country_data[year_columns].values.flatten(), label=country)

plt.title("Inflation Trends (1960-2024)")
plt.xlabel("Year")
plt.ylabel("Inflation Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

2. Inflation Volatility Ranking(Inflation volatility refers to how much inflation rates fluctuate)

```python
df_merge['volatility'] = df_merge[year_columns].std(axis=1)
top_volatile = df_merge.sort_values(by='volatility', ascending=False).head(10)
least_volatile = df_merge.sort_values(by='volatility', ascending=True).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='volatility', y='Country Name', data=top_volatile)
plt.title("Top 10 Countries by Inflation Volatility")
plt.show()
```

3.Regional/Income Group

```python

# Filter only relevant columns: region, income group, and yearly inflation data
df_region_income = df_merge[["Region", "IncomeGroup"] + year_columns].copy()

# Replace empty strings or NaNs in Region and IncomeGroup if any
df_region_income["Region"].replace("", "Unknown", inplace=True)
df_region_income["IncomeGroup"].replace("", "Unknown", inplace=True)

# Calculate mean inflation per year grouped by Region
region_grouped = df_region_income.groupby("Region")[year_columns].mean().T

# Calculate mean inflation per year grouped by IncomeGroup
income_grouped = df_region_income.groupby("IncomeGroup")[year_columns].mean().T

# Plot mean inflation trends by Region
plt.figure(figsize=(14,6))
for region in region_grouped.columns:
    plt.plot(region_grouped.index.astype(int), region_grouped[region], label=region)
plt.title("Mean Inflation Trend by Region (1960-2024)")
plt.xlabel("Year")
plt.ylabel("Inflation Rate (%)")
plt.legend()
plt.grid(True)
plt.show()

# Plot mean inflation trends by IncomeGroup
plt.figure(figsize=(14,6))
for group in income_grouped.columns:
    plt.plot(income_grouped.index.astype(int), income_grouped[group], label=group)
plt.title("Mean Inflation Trend by Income Group (1960-2024)")
plt.xlabel("Year")
plt.ylabel("Inflation Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Melt dataframe to long format for last 10 years for boxplot
recent_years = year_columns[-10:]
df_long = df_region_income.melt(
    id_vars=["IncomeGroup"],
    value_vars=recent_years,
    var_name="Year",
    value_name="Inflation"
)

# Drop missing values (NaNs) if any
df_long = df_long.dropna(subset=["Inflation"])

# Create boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df_long, x="IncomeGroup", y="Inflation")
plt.title("Inflation Distribution by Income Group (Last 10 Years)")
plt.ylabel("Inflation Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

III.Inflation Volatility and Hyperinflation Case Study

```python
# Compute volatility for all countries (std deviation over all years)
df_merge['volatility'] = df_merge[year_columns].std(axis=1)

# Top 5 most volatile countries (usually hyperinflation cases)
top_volatile = df_merge.sort_values(by='volatility', ascending=False).head(5)

print("Top 5 Most Volatile Inflation Countries:")
print(top_volatile[['Country Name', 'volatility']])

# Plot inflation trends for top 5 volatile countries
plt.figure(figsize=(16,8))
for idx, row in top_volatile.iterrows():
    plt.plot([int(y) for y in year_columns], row[year_columns].values, label=row['Country Name'])

plt.title("Top 5 Most Volatile Inflation Countries (1960-2024)")
plt.xlabel("Year")
plt.ylabel("Inflation Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

Insights: Hyperinflation and Volatility: Economic Case Studies

The countries with the most volatile inflation (Congo, Dem. Rep.; Bolivia; Peru; Angola; Ukraine) experienced periods of hyperinflation, visible as extreme spikes in the chart above.  
- **Congo, Dem. Rep.** saw sustained hyperinflation in the early 1990s driven by political instability and monetary collapse.
- **Bolivia** and **Peru** faced similar runaway inflation in the 1980s and 1990s, largely due to currency devaluation and fiscal crises.
- **Angola**'s post-civil war transition period resulted in explosive price levels.
- **Ukraine** suffered hyperinflation after the USSR breakup as its monetary system transitioned in the 1990s.

These cases are useful for illustrating how external shocks, political change, and fiscal mismanagement can destabilize economies, and the results are observable even decades later in the global inflation data.


IV.Simple Time Series Forecasting

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Select a country (India as example) and prepare time series data
country_to_forecast = "India"
country_data = df_merge[df_merge['Country Name'] == country_to_forecast][year_columns].T
country_data = country_data.dropna().astype(float)
country_data.reset_index(inplace=True)
country_data.rename(columns={country_data.columns[1]: 'Inflation'}, inplace=True)
country_data['Year'] = country_data['index'].astype(int)

# Simple Linear Regression forecasting last 20 years
X = country_data['Year'].values.reshape(-1, 1)
y = country_data['Inflation'].values

model = LinearRegression()
model.fit(X, y)
next_year = np.array([[country_data['Year'].max() + 1]])
forecast = model.predict(next_year)[0]

# Plot actual and predicted
plt.figure(figsize=(10,6))
plt.plot(country_data['Year'], country_data['Inflation'], marker='o', label='Actual')
plt.plot(
    [country_data['Year'].max(), country_data['Year'].max() + 1],
    [country_data['Inflation'].iloc[-1], forecast],
    'r--o', label='Forecast'
)
plt.title(f"{country_to_forecast} Inflation: Actual and Linear Regression Forecast")
plt.xlabel("Year")
plt.ylabel("Inflation Rate (%)")
plt.legend()
plt.grid()
plt.show()

print(f"Predicted inflation for {country_to_forecast} in {int(next_year[0][0])}: {forecast:.2f}%")
```

V.Streamlit Dashboard

```python
import streamlit as st
import pandas as pd
import cufflinks as cf
import plotly.graph_objs as go

cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)

@st.cache_data
def load_data():
    df = pd.read_csv(r"API_FP.CPI.TOTL.ZG_DS2_en_csv_v2_713244.csv", skiprows=4)
    meta_country = pd.read_csv(r"Metadata_Country_API_FP.CPI.TOTL.ZG_DS2_en_csv_v2_713244.csv")
    df.columns = df.columns.str.strip()
    df = df[df["Indicator Code"] == "FP.CPI.TOTL.ZG"]
    df_merge = df.merge(meta_country, on="Country Code", how="left")
    return df_merge

df_merge = load_data()

st.sidebar.title("Inflation Data Explorer")
selected_region = st.sidebar.selectbox("Select Region", ["All"] + sorted(df_merge["Region"].dropna().unique().tolist()))
selected_income = st.sidebar.selectbox("Select Income Group", ["All"] + sorted(df_merge["IncomeGroup"].dropna().unique().tolist()))
selected_years = st.sidebar.slider("Select Year Range", 1960, 2024, (2000, 2020))

df_filtered = df_merge.copy()
if selected_region != "All":
    df_filtered = df_filtered[df_filtered["Region"] == selected_region]
if selected_income != "All":
    df_filtered = df_filtered[df_filtered["IncomeGroup"] == selected_income]

year_cols = [str(y) for y in range(selected_years[0], selected_years[1] + 1)]
df_filtered = df_filtered[["Country Name"] + year_cols]

st.title("Inflation Trends Explorer")
st.write(f"Showing inflation from {selected_years[0]} to {selected_years[1]}")

df_plot = df_filtered.set_index("Country Name").T
df_plot.index = df_plot.index.astype(int)
df_plot = df_plot.apply(pd.to_numeric, errors='coerce')  # Ensure numeric

fig = go.Figure()
for country in df_plot.columns:
    fig.add_trace(go.Scatter(
        x=df_plot.index,
        y=df_plot[country],
        mode='lines+markers',
        name=country
    ))

fig.update_layout(
    title='Inflation Trends by Country',
    xaxis_title='Year',
    yaxis_title='Inflation Rate (%)',
    template='plotly_white',
    height=600
)

st.plotly_chart(fig, use_container_width=True)
```
